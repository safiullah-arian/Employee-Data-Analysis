```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("DIKW_HW_Activity.csv")
```

```python
df.head()
```

```python
# What are the different types of data present (e.g., categorical, numerical)?
df.dtypes
```

```python
# Calculate the average salary for each department.
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()
avg_salary_by_dept
```

```python
# Find the number of employees in each country.
employee_count_by_country = df["Country"].value_counts()
employee_count_by_country
```

```python
# Determine the average performance score for all employees
avg_performance_score = df["PerformanceScore"].mean()
avg_performance_score
```

```python
#Department with the highest average performance score
highest_perf_dept = df.groupby("Department")["PerformanceScore"].mean().idxmax()
highest_avg_perf_score = df.groupby("Department")["PerformanceScore"].mean().max()
highest_perf_dept, highest_avg_perf_score
```

```python
# Countries with the highest job satisfaction score
highest_satisfaction_country = df.groupby("Country")["JobSatisfaction"].mean().idxmax()
highest_satisfaction_score = df.groupby("Country")["JobSatisfaction"].mean().max()
highest_satisfaction_country, highest_satisfaction_score
```

```python
# Department with the lowest job satisfaction score
lowest_satisfaction_dept = df.groupby("Department")["JobSatisfaction"].mean().idxmin()
lowest_satisfaction_score = df.groupby("Department")["JobSatisfaction"].mean().min()
lowest_satisfaction_dept, lowest_satisfaction_score
```

```python
# Average salary by gender within each department
salary_by_gender = df.groupby(["Department", "Gender"])["Salary"].mean()
salary_by_gender
```

```python
# Plot average salary by department
avg_salary_by_dept.plot(kind='bar', title="Average Salary by Department", ylabel="Salary ($)")
plt.show()
```

```python
# Plot average performance score by department
df.groupby("Department")["PerformanceScore"].mean().plot(kind='bar', title="Average Performance Score by Department", ylabel="Performance Score")
plt.show()
```

```python
# Plot average job satisfaction by country
df.groupby("Country")["JobSatisfaction"].mean().plot(kind='bar', title="Average Job Satisfaction by Country", ylabel="Job Satisfaction Score")
plt.show()
```
